```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina' # 폰트가 깨끗하게 보이도록 설정

!apt install fonts-nanum # 나눔 폰트 설치
import matplotlib.font_manager as fm
fontpath = '/content/drive/MyDrive/NaturalLanguageProcessing/NanumMyeongjo.ttf'
font = fm.FontProperties(fname=fontpath, size=9)

# 기본 글꼴 변경
import matplotlib as mpl
mpl.font_manager._rebuild()
mpl.pyplot.rc('font', family='NanumMyeongjo')
```

```python
with open('/content/drive/MyDrive/NaturalLanguageProcessing/pytest/윤동주시.txt', 'r', encoding = 'utf8') as f:
  b=f.read()

print(b)
```

```python
sample_data = rhinoMorph.onlyMorph_list(rn, b, pos=['NNG', 'NNP', 'VV', 'VA', 'XR', 'IC', 'MM', 'MAG', 'MAJ'], eomi=True)

wordInfo = Counter(sample_data)
print('wordInfo:', wordInfo)
```

```python
sorted_keys = sorted(wordInfo, key=wordInfo.get, reverse=True)
sorted_values = sorted(wordInfo.values(), reverse=True)

import matplotlib.pyplot as plt
plt.bar(range(10), sorted_values[:10])
plt.xticks(range(10), sorted_keys[:10])
plt.show()
```

```python
!pip install wordcloud # 워드클라우드 패키지 설치

from wordcloud import WordCloud

cloud = WordCloud(font_path=fontpath, width=800, height=600, background_color = 'white').generate(" ".join(wordInfo))
plt.imshow(cloud, interpolation='bilinear') # 글자를 더 부드럽게 나오게 한다
plt.axis('off') # 축의 위치 정보 off
plt.show()
```
